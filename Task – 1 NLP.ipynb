# Task 1: Conceptual Understanding

## 1. Difference between "Love" and "love" in NLP

In Natural Language Processing (NLP), "Love" and "love" are treated as different tokens because NLP models are case-sensitive by default. "Love" may appear at the beginning of a sentence or as a proper noun, while "love" is usually used as a verb or noun in general context.

However, during preprocessing, we often convert all text to lowercase to maintain consistency and reduce redundancy. This process is called **case normalization**, and it ensures that both "Love" and "love" are treated as the same word.



## 2. What happens if stopwords are not removed?

Stopwords are common words such as "is", "the", "and", "in", etc. If stopwords are not removed:

* The dataset becomes larger and noisier
* It increases computational complexity
* It may reduce model performance by adding irrelevant information
* Important keywords may get less importance due to frequency of stopwords

However, in some cases, stopwords carry meaning and should not always be removed.



## 3. Real-world scenarios where removing stopwords can be harmful

### Scenario 1: Sentiment Analysis

In sentences like:

* "I do not like this product"

Removing stopwords like "not" changes the meaning completely:

* "I like this product" (incorrect sentiment)

### Scenario 2: Question Answering Systems

Example:

* "What is the capital of India?"

Removing stopwords may result in:

* "capital India"

This can affect understanding of the question and lead to incorrect answers.



## 4. Difference between stemming and lemmatization

### Stemming

* Reduces words to their root form by cutting off suffixes
* It is faster but less accurate
* Example:

  * "running" → "run"
  * "studies" → "studi"

### Lemmatization

* Converts words to their meaningful base form using vocabulary and grammar rules
* It is slower but more accurate
* Example:

  * "running" → "run"
  * "studies" → "study"

### Key Difference:

Stemming is a **rule-based shortcut**, while lemmatization is **context-aware and linguistically correct**.


***_Task 2:Build Advanced Preprocessing Function_***


```python
import re

def preprocess_text(text):
    
    # 1. Convert to lowercase
    text = text.lower()
    
    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # 3. Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)
    
    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 5. Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    
    # 6. Remove special characters / punctuation
    text = re.sub(r'[^a-z\s]', '', text)
    
    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 8. Remove very short tokens (length ≤ 2)
    words = text.split()
    cleaned_words = [
        word for word in words 
        if len(word) > 2 or word in ['no', 'not']
    ]
    
    return " ".join(cleaned_words)
```

```python
text = "WOW!!! This is sooo goooood!!! Visit http://example.com now 123"
print(preprocess_text(text))
```

**_Task 3: Stress Testing_**

```python
# Sample Inputs
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://amazon.in now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Stress Testing
for i, sentence in enumerate(test_sentences, 1):
    
    cleaned_sentence = preprocess_text(sentence)
    tokens = cleaned_sentence.split()
    
    print(f"--- Sentence {i} ---")
    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print()
```

***_Task 4: Token Analytics_***

```python
# Token Analytics

for i, sentence in enumerate(test_sentences, 1):
    
    cleaned_sentence = preprocess_text(sentence)
    tokens = cleaned_sentence.split()
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    
    if total_tokens > 0:
        avg_token_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_token_length = 0
    
    print(f"--- Sentence {i} ---")
    print("Cleaned Sentence :", cleaned_sentence)
    print("Total Tokens     :", total_tokens)
    print("Unique Tokens    :", unique_tokens)
    print("Avg Token Length :", round(avg_token_length, 2))
    print()
```

***_Task 5: Frequency Analysis_***

```python
from collections import Counter

# Collect all tokens from all sentences
all_tokens = []

for sentence in test_sentences:
    cleaned_sentence = preprocess_text(sentence)
    tokens = cleaned_sentence.split()
    all_tokens.extend(tokens)

# Create frequency counter
word_freq = Counter(all_tokens)

# Top 10 most frequent words
top_10 = word_freq.most_common(10)

# Top 5 least frequent words
least_5 = word_freq.most_common()[:-6:-1]

print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(f"{word} : {count}")

print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(f"{word} : {count}")
```

***_Task 6: Build Full Pipeline_***

```python
def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        cleaned = preprocess_text(text)
        tokens = cleaned.split()
        
        clean_sentences.append(cleaned)
        all_tokens.extend(tokens)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)

print("Clean Sentences:")
for sentence in result["clean_sentences"]:
    print("-", sentence)

print("\nAll Tokens:")
print(result["tokens"])
```

***_Task 7: Error Handling_***

```python
import re

def preprocess_text(text):
    
    # Handle empty input
    if not text or not isinstance(text, str):
        return ""
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Handle repeated characters (fix)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    
    # Remove special characters (including emojis)
    text = re.sub(r'[^a-z\s]', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # If nothing remains after cleaning
    if not text:
        return ""
    
    # Remove short words
    words = text.split()
    cleaned_words = [
        word for word in words 
        if len(word) > 2 or word in ['no', 'not']
    ]
    
    return " ".join(cleaned_words)
```

```python
test_cases = [
    "",
    "😍😍🔥🔥",
    "123456789",
]

for i, case in enumerate(test_cases, 1):
    print(f"Test Case {i}:")
    print("Input  :", case)
    print("Output :", preprocess_text(case))
    print("-" * 40)
```
